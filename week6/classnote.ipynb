```python
const numbers = [1, 2, 3, 4, 5];

const users = [
    {name: 'Alice', age: 25, active: false},
    {name: 'Bob', age: 30, active: true},
    {name: 'Charlie', age: 35, active: true},
    {name: 'David', age: 40, active: false},
    {name: 'Eve', age: 45, active: true}
]


// map
const numberDoubled = numbers.map((num) => {
    return num * 2;
});

// filter
const oddNumbers = numbers.filter((num) => {
    return (num % 2) === 1; // % modulus. Use === since you want to get the 'exact' 1 as exact value.
});

// ex.2
const activeUsers = users.filter((user) => {
    return user.active;
});

const inactiveUsers = users.filter((user) => {
    return !user.active;
});

// reduce: reduce the array to 1 item
const averageUserAge = users.reduce((total, user) => {
    total += user.age;
}, 0) / users.length;

```
